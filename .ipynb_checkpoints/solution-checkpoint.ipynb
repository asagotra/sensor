```python
# Load libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import TruncatedSVD
from scipy.sparse import csr_matrix
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
```

```python
# data load and preprocessing using pandas
#References: https://pandas.pydata.org/

df= pd.read_csv("task_data.csv")

# Extract names of each column 
headers = df.columns.values
names = headers[2:]

# Extract features 
df_to_array = df.values
X = df_to_array[:,2:]
print ("Features shape is {}".format(X.shape))

# Extract labels 
Y = df['class_label'].values
print ("Labels shape is {}".format(Y.shape))
```

```python

# References: https://scikit-learn.org/stable/
# Refereces: The Elements of Statistical Learning: Data Mining, Inference, and Prediction, Second Edition (Springer Series in Statistics)

# Random Forest 
def random_forest():
    model = RandomForestRegressor()
    model.fit(X,Y)
    return model
rf=random_forest()

# Extra tree classifier
def extra_tree():
    model = ExtraTreesClassifier()
    model.fit(X, Y)
    return model
etc = extra_tree()

# gradient boosting classifier
def xgb_classifier():
    model = XGBClassifier()
    model.fit(X, Y)
    return model
xgb = xgb_classifier()


```

```python
# Calculation by using Random Forest
# Sort in descending order
rank_rf = sorted(zip( rf.feature_importances_, names), reverse=True)
print("Predictive Scores and Features:")
for scores_rf in rank_rf: 
    print(scores_rf)
# save data as .csv file     
import csv
res =rank_rf
csvfile = "rf_features.csv"
with open(csvfile, "w") as output:
    writer = csv.writer(output, lineterminator='\n')
    writer.writerows(res)
    
# Visualizations and Save the labels and their importance scores separately

# Reference: http://stackoverflow.com/questions/13925251/python-bar-plot-from-list-of-tuples

score_rf = list(zip(*rank_rf))[0]
sensors = list(zip(*rank_rf))[1]

x_pos = np.arange(len(sensors)) 
fig = plt.figure(figsize=(11, 6))
sns.barplot(x_pos, score_rf, color="green")
plt.xticks(x_pos, sensors) 
plt.ylabel('Importance Scores')
plt.title("Random Forest")
plt.savefig("rf_features.png")
plt.show()
```

```python
# Calculation by using ExtraTreeClassifier
# Sort in descending order
rank_etc = sorted(zip(etc.feature_importances_, names), reverse=True)
print("Predictive Scores and Features:")
for scores_etc in rank_etc: 
    print(scores_etc)

# save data as .csv file     
import csv
res =rank_etc
csvfile = "etc_features.csv"
with open(csvfile, "w") as output:
    writer = csv.writer(output, lineterminator='\n')
    writer.writerows(res)
    
#visualization and Save the labels and their importance scores separately

score_etc = list(zip(*rank_etc))[0]
sensors = list(zip(*rank_etc))[1]

x_pos = np.arange(len(sensors)) 
fig = plt.figure(figsize=(11, 6))
sns.barplot(x_pos, score_etc, color="green")
plt.xticks(x_pos, sensors) 
plt.ylabel('Importance Scores')
plt.title("ExtraTreeClassifier")
plt.savefig("etc_features.png")
plt.show()
```

```python
#calculation by using XGBClassifier
#Sort in descending order
rank_xgb = sorted(zip(xgb.feature_importances_, names), reverse=True)
print("Predictive Scores and Features:")
for scores_xgb in rank_xgb: 
    print(scores_xgb)

# save data as .csv file     
import csv
res =rank_xgb
csvfile = "xgb_features.csv"
with open(csvfile, "w") as output:
    writer = csv.writer(output, lineterminator='\n')
    writer.writerows(res)
    
#visualization and Save the labels and their importance scores separately

score_xgb = list(zip(*rank_xgb))[0]
sensors = list(zip(*rank_xgb))[1]

x_pos = np.arange(len(sensors)) 
fig = plt.figure(figsize=(11, 6))
sns.barplot(x_pos, score_xgb, color="green")
plt.xticks(x_pos, sensors) 
plt.ylabel('Importance Scores')
plt.title("XGBClassifier")
plt.savefig("xgb_features.png")
plt.show()

# if you want to see the f-score of the features then uncommented these two lines
#from xgboost import plot_importance
#plot_importance(xgb)
```

```python
# Heatmap of Correlation Matrix 
# References: https://seaborn.pydata.org/examples/many_pairwise_correlations.html

df_new = df.drop(['sample index', 'class_label'], axis=1)

#df_new= df_new.set_index(['class_label', 'sensor'])

corrmat = df_new.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(15,15))

#plot heat map
g=sns.heatmap(df[top_corr_features].corr(), annot=True, cmap="coolwarm")
plt.savefig("corr_matrix.png")
```

```python
# Calcultion by Using Random Foest Classifier

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

#Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)

#Import scikit-learn metrics module for accuracy calculation

from sklearn import metrics

# Model Accuracy, how often is the classifier correct?

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# sorting the features

rank_clf = sorted(zip(clf.feature_importances_, names), reverse=True)
print("Predictive Scores and Features:")
for scores_clf in rank_clf: 
    print(scores_clf)

# save data as .csv file     
import csv
res =rank_clf
csvfile = "clf_features.csv"
with open(csvfile, "w") as output:
    writer = csv.writer(output, lineterminator='\n')
    writer.writerows(res)
    
#visualization and Save the labels and their importance scores separately

score_clf = list(zip(*rank_clf))[0]
sensors = list(zip(*rank_clf))[1]

x_pos = np.arange(len(sensors)) 
fig = plt.figure(figsize=(11, 6))
sns.barplot(x_pos, score_clf, color="green")
plt.xticks(x_pos, sensors) 
plt.ylabel('Importance Scores')
plt.title("RandomForestClassifier")
plt.savefig("clf_features.png")
plt.show()
```

```python
# scalability  with respect to number of features

from sklearn.metrics import accuracy_score
from sklearn.feature_selection import SelectFromModel
from numpy import sort
# fit model on all training data
model = XGBClassifier()
model.fit(X_train, y_train)
# make predictions for test data and evaluate
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
# Fit model using each importance as a threshold
thresholds = sort(model.feature_importances_)
for thresh in thresholds:
	# select features using threshold
	selection = SelectFromModel(model, threshold=thresh, prefit=True)
	select_X_train = selection.transform(X_train)
	# train model
	selection_model = XGBClassifier()
	selection_model.fit(select_X_train, y_train)
	# eval model
	select_X_test = selection.transform(X_test)
	y_pred = selection_model.predict(select_X_test)
	predictions = [round(value) for value in y_pred]
	accuracy = accuracy_score(y_test, predictions)
	print("Thresh=%.5f, n=%d, Accuracy: %.2f%%" % (thresh, select_X_train.shape[1], accuracy*100))
```
